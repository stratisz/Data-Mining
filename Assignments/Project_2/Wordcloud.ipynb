```python
import wordcloud
import os

import matplotlib.pyplot as plt
```

# Business

```python
file_list_business = []

for root,dir,files in os.walk("./fulltext/data/business"):

    for f in files:
        
        x = os.path.join(root,f)
        x = x.replace("\\","/")
        file_list_business.append(x)
```

```python
text = ""

for f in file_list_business:

    fp = open(f,"r" , encoding='unicode_escape')
    
    line_list = fp.readlines()
    
    x1 = ''.join(line_list[1:]) #skip title
    x1 = x1.replace('\n',' ')
    
    
    text = text + x1
    fp.close()   
```

```python
w1 = wordcloud.WordCloud(width=3000,height=3000, stopwords=wordcloud.STOPWORDS).generate(text)

#or 1000,500

plt.imshow(w1,interpolation='bilinear')
plt.axis='off'
plt.show()
```
